# World map - Transactions locations

```python
import folium
from geopy.geocoders import Nominatim
#Geopy nos permite convertir direcciones o nomobre de lugares en coordenadas geograficas
```

```python
unique_locations = df['Location'].unique()
location_coords = {}
```

```python
geolocator = Nominatim(user_agent="location_mapper")
```

```python
for location in unique_locations:
    try:
        loc = geolocator.geocode(location)
        if loc:
            location_coords[location] = (loc.latitude, loc.longitude)
        else:
            print(f"Coordinates not found for {location}")
    except Exception as e:
        print(f"Error fetching coordinates for {location} : {e}")
```

```python
df['Coordinates'] = df['Location'].map(location_coords)
```

```python
initial_coords = list(location_coords.values())[0] if location_coords else [0, 0]
mymap = folium.Map(location = initial_coords, zoom_start=5)
```

```python
for _, row in df.iterrows():
    if row['Coordinates']:
        folium.Marker(
            location=row['Coordinates'],
            popup=f"TransactionID: {row['TransactionID']}<br>Amount: ${row['TransactionAmount']}",
            tooltip=row['Location']
        ).add_to(mymap)
```

# Detecting total potential frauds | K-means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df[["TransactionAmount", "CustomerAge"]]  #Storage the variables "TransactionAmount" & "CustomerAge" in the variables "X"
```

***ESTANDARIZACIÃ“N***

La estandarizaciÃ³n es una tÃ©cnica de preprocesamiento de datos muy comÃºn en machine learning. Se realiza por varias razones:
*   **Escalado de caracterÃ­sticas:** Al tener caracterÃ­sticas con diferentes escalas (por ejemplo, TransactionAmount puede variar en miles, mientras que CustomerAge varÃ­a en dÃ©cadas), los algoritmos de aprendizaje automÃ¡tico pueden dar mÃ¡s importancia a las caracterÃ­sticas con mayor rango numÃ©rico. La estandarizaciÃ³n asegura que todas las caracterÃ­sticas tengan la misma escala, lo que permite que el algoritmo aprenda de manera mÃ¡s justa.
*   **Mejora de la convergencia de algoritmos:** Muchos algoritmos de aprendizaje automÃ¡tico, como K-means, funcionan mejor con datos estandarizados. La estandarizaciÃ³n puede ayudar a acelerar la convergencia del algoritmo y mejorar su precisiÃ³n.
*  **Centrado de datos:** Al restar la media, los datos se centran en el origen, lo que puede ser Ãºtil para algunos algoritmos.








```python
#Data Standardization
scaler=StandardScaler()   #Create an instance of StandardScaler
X_scaled=scaler.fit_transform(X)  #Esta parte del cÃ³digo calcula los parÃ¡metros necesarios para la estandarizaciÃ³n de los datos almacenados en "X"
```

```python
#K-Means Clustering

#The line below creates an instance of KMeans and stablish 3 parameters:
#n_cluster: It indicates that it has to create 3 clusters
#random_state: it stablish the "semilla aleatoria"
Kmeans=KMeans(n_clusters=3, random_state=42)

#the line below group the data and store them in 3 different labels (groups)
Kmeans_labels=Kmeans.fit_predict(X_scaled)

#Add lables to the dataset
df["Kmeans_Cluster"]=Kmeans_labels
```

Aquellos valores que esten mas alejados de su centroide son outliers, por lo tanto, pueden ser fraides bancarios.
Se calcula el percentil 95 de las distancias entre los puntos de datos y sus centroides. Esto significa que el 95% de los puntos estÃ¡n a una distancia menor o igual a este valor.

```python
import numpy as np

# Calculate distance of each point to its assigned cluster centroid
distances = np.linalg.norm(X_scaled - Kmeans.cluster_centers_[Kmeans_labels], axis=1)
# Define a threshold (e.g., top 5% farthest from centroids)
threshold = np.percentile(distances, 95)  # Change this value as needed
```

Se crea una nueva columna llamada "Potential_Fraud, esta columna tendra valores booleanos, si un punto cumple con la condiciÃ³n, sera considerado fraude potencial.

```python
# Flag points above the threshold as potential frauds
df['Potential_Fraud'] = distances > threshold
```

Creamos dos DF, uno con los fraudes y uno con aquellas transacciones que no son consideradas fraudes.

En esta linea de codigo se usa ( ~ ), el cual es una negaciÃ³n logica, cambia el valor del booleano que tengamos, un True lo vuelve False y viceversa.
Los DF seran creados con aquellos valores que tenga Booleanos True.

```python
#Separate fraudulent and non fraudulent transactions
frauds=df[df['Potential_Fraud']]
non_frauds=df[~df['Potential_Fraud']]
```

```python
# Plot with frauds highlighted
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=Kmeans_labels, palette='viridis', s=60, alpha=0.5)
plt.scatter(Kmeans.cluster_centers_[:, 0], Kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids')
plt.scatter(X_scaled[distances > threshold, 0], X_scaled[distances > threshold, 1],
            color='black', s=50, label='Potential Frauds', marker='X')
plt.title('K-means Clustering with Potential Frauds Highlighted')
plt.xlabel('Scaled Amount')
plt.ylabel('Scaled Age')
plt.legend()
```

```python
# Output the number of frauds detected
print(f"Number of potential frauds detected: {len(frauds)}")
```

DBSCAN Clustering on Transactions

```python
from sklearn.cluster import DBSCAN

# Adjusted DBSCAN clustering with a refined eps value
dbscan = DBSCAN(eps=0.3, min_samples=5)  # Adjust eps based on k-distance elbow
dbscan_labels = dbscan.fit_predict(X_scaled)

# Add labels to the dataset
df['DBSCAN_Cluster'] = dbscan_labels

# Map the cluster labels to descriptive names
label_mapping = {-1: 'Fraud', 0: 'Normal', 1: 'Suspicious Group 1', 2: 'Suspicious Group 2'}
df['DBSCAN_Cluster'] = df['DBSCAN_Cluster'].map(label_mapping)

# Plotting with updated labels
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=df['DBSCAN_Cluster'], palette='coolwarm', s=60)
plt.title('DBSCAN Clustering on Transactions')
plt.xlabel('Scaled Amount')
plt.ylabel('Scaled Age')
plt.legend(title='Cluster')
plt.show()
```
